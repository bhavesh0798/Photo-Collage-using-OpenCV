```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
t_left = cv2.imread("top_left.jpg")
t_right = cv2.imread("top_right.jpg")
b_left = cv2.imread("bottom_left.jpg")
b_right = cv2.imread("bottom_right.jpg")
c = cv2.imread("center.jpeg")

#converting to rgb from bgr
t_left = cv2.cvtColor(t_left,cv2.COLOR_BGR2RGB)
t_right = cv2.cvtColor(t_right,cv2.COLOR_BGR2RGB)
b_left = cv2.cvtColor(b_left,cv2.COLOR_BGR2RGB)
b_right = cv2.cvtColor(b_right,cv2.COLOR_BGR2RGB)
c = cv2.cvtColor(c,cv2.COLOR_BGR2RGB)

t_left = cv2.resize(t_left,(200,200))
t_right = cv2.resize(t_right,(200,200))
b_left = cv2.resize(b_left,(200,200))
b_right = cv2.resize(b_right,(200,200))
c = cv2.resize(c,(200,200))
```

```python
#lets see the images
plt.imshow(t_left)
plt.show()
```

```python
#add borders to all images
t_left = cv2.copyMakeBorder(t_left,10,10,10,0,0)
t_right = cv2.copyMakeBorder(t_right,10,10,10,10,0)
b_left = cv2.copyMakeBorder(b_left,0,10,10,0,0)
b_right = cv2.copyMakeBorder(b_right,0,10,10,10,0)
c = cv2.copyMakeBorder(c,10,10,10,10,0)
```

```python
plt.imshow(t_left)
plt.show()
```

```python
plt.imshow(c)
plt.show()
```

```python
#create a matrix of (430,430,3)
matrix = np.zeros((430,430,3),int)
```

```python
matrix[0:220,0:210] = t_left
matrix[0:220,210:430] = t_right
matrix[220:430,0:210] = b_left
matrix[220:430,210:430] = b_right
matrix[110:330,110:330] = c
```

```python
plt.imshow(matrix)
plt.axis("off")
plt.show()
```

```python
print(type(matrix[0][0][0]))
```

```python
mat = matrix.reshape(184900,3)
```
